```python
import pandas as pd
import os
```

This notebook updates the main config to include `deg_path`.
These are significantly differentially expressed genes as analyzed by Rasha Aljelaify.

```python
INPUT_DIR = "original_data/ST_marker_genes/ST-markers"
CONFIG = "classify/main_config.csv"
deg_files = os.listdir(INPUT_DIR)
```

```python
deg_paths_dict = {filename.split('-')[1] : INPUT_DIR + "/" + filename 
                  for filename in deg_files}

deg_paths_dict['86A'] = "original_data/ST_marker_genes/ST-markers/S61-S55-86A-NT1_markers.csv"
deg_paths_dict['86B'] = "original_data/ST_marker_genes/ST-markers/S61-S55-86B-NT2_markers.csv"

config = pd.read_csv(CONFIG)
# Get 33A from V10F03-033_A
easy_read_ids = config.loc[:, 'sample_id'].map(lambda x:  x[8:10] + x[11])
# Make new col with deg_paths
config.loc[:, 'deg_path'] = easy_read_ids.map(deg_paths_dict)
# Save to main_config
config.to_csv(CONFIG)
```
